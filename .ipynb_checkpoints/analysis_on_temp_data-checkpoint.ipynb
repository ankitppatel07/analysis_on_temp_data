```python
import pandas as pd
import matplotlib.pyplot as plt

df_Country = pd.read_excel("data\GlobalLandTempByCountry.xlsx", sheet_name='Sheet1')
df_Major_City = pd.read_excel("data\GlobalLandTempByMajorCity.xlsx", sheet_name='Sheet1')
df_Global = pd.read_excel("data\GlobalTemp.xlsx", sheet_name='Sheet1')

print(df_Country.columns)
```

```python
pd.set_option('display.width',100)
df_Country.head(3)
```

```python
df_Major_City.head(3)
```

```python
df_Global.head(3)
```

### - Calculating moving averages of temperature data rather than the yearly averages in order to smooth out the lines when plotting a line graph, making trends more observable

```python
df_Country['12_Month_MA'] = df_Country['AverageTemperature'].rolling(window=12).mean()
df_Country.head(13)
```

```python
df_Major_City['12_Month_MA'] = df_Major_City['AverageTemperature'].rolling(window=12).mean()
df_Major_City.head(13)
```

```python
df_Global['12_Month_MA'] = df_Global['LandAverageTemperature'].rolling(window=12).mean()
df_Global.head(13)
```

### - Comparing the global temperature with the temperature of the city Bombay(Mumbai), India (using a line graph):


```python
bombay_city_temp_values = df_Major_City['12_Month_MA'][df_Major_City['City']=='Bombay']
city_date_values = df_Major_City['Date'][df_Major_City['City']=='Bombay']
global_temp_values = df_Global['12_Month_MA']

plt.figure(figsize=(15,6))
plt.plot(city_date_values, bombay_city_temp_values, color = 'blue', linewidth=3, label='Temp of Bombay')
plt.plot(df_Global['Date'], global_temp_values, color = 'red', linewidth=3, label='Global Temp')
plt.title("Comparing the Global temperature with the temperature of Bombay(Mumbai) from 1900 till date", fontsize=15)
plt.xlabel("Timeline", fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("Temperature(degree celsius)", fontsize=15)
plt.yticks(fontsize=12)
plt.axis(["1900-01-01","2013-09-01",0,50])
plt.legend(fontsize=12)
```

### - Comparing the temperature of USA and India from 1900 till date (using a line graph):

```python
india_temp_values = df_Country['12_Month_MA'][df_Country['Country']=='India']
usa_temp_values = df_Country['12_Month_MA'][df_Country['Country']=='United States']
india_date_values = df_Country['Date'][df_Country['Country']=='India']
usa_date_values = df_Country['Date'][df_Country['Country']=='United States']

plt.figure(figsize=(15,6))
plt.plot(india_date_values, india_temp_values, color = 'blue', linewidth=3, label='Temp of India')
plt.plot(usa_date_values, usa_temp_values, color = 'red', linewidth=3, label='Temp of USA')
plt.title("Comparing the temperature of USA and India from 1900 till date", fontsize=15)
plt.xlabel("Timeline", fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("Temperature(degree celsius)", fontsize=15)
plt.yticks(fontsize=12)
plt.axis(["1900-01-01","2013-09-01",0,50])
plt.legend(fontsize=12)
```

***
## Insights:
- The city of Bombay(Mumbai) is hotter on average when compared to the global average temperature(degree celsius) 
  and the difference has been consistent for the past 100 years.
- There is negligible increase in the temperature of the city of Bombay(Mumbai) and the Global temperature when we consider   the temperature data for the past 100 years. The world is getting very slightly hotter.
- We can see almost similar trends when we compare the countries USA and India. Though USA is much cooler than India, there   is a very slight increase in the temperature of both the countries when we consider the temperature data for the past 100   years.
